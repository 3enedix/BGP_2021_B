### Landuse: Percentage of of total area of the Netherlands per landcover type

```python
perc_land_cover = pd.DataFrame()
for lcover in ['built_up', 'dry_natural_terrain', 'greenhouse_horticulture', 'railway', 'nature_reserves', 'company_site', 'wet_natural_terrain', 'airport',
               'forest', 'coastal_water', 'semi_built', 'agriculture', 'recreation', 'water', 'highway']:
    perc_land_cover.at[0,lcover] = round(data[lcover].sum()/(len(data)*1000*1000),2)
perc_land_cover
```

### Normalise Data

```python
data_norm = pd.DataFrame()
for (colname, coldata) in data.iteritems(): # go through columns of dataframe
    if colname == 'obsdate' \
    or colname == 'block' \
    or colname == 'observer_intensity' \
    or colname == 'mdate' \
    or colname == 'holiday' \
    or colname == 'work_day':
        data_norm[colname] = data[colname]
    else:
        data_norm[colname] = (data[colname] - data[colname].mean()) / (data[colname].max() - data[colname].min())
        
```

```python
data_norm.head()
```

```python
data_norm.describe()
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.hist(data_norm['observer_intensity'][data_norm['observer_intensity'] <= 10], bins=10)
plt.title('Observer Intensity')
```

```python
fig, ax = plt.subplots(figsize=(10,5))
data_norm['observer_intensity'][data_norm['observer_intensity'] <= 10].plot(kind='kde')
plt.title('Observer Intensity: Probability Density Function')
```

### Histograms for some of the parameters

### Correlations

```python
corr = data_norm.corr()
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
cax = ax.matshow(corr,cmap='coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(data_norm.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(data_norm.columns)
ax.set_yticklabels(data_norm.columns)
plt.show()
```

```python
import seaborn
df_seaborn_heatmap=data_norm[['observer_intensity', 'precipitation', 'temperature', 'municipality', 'roadlength', 'nature_reserves', 'built_up', 'dry_natural_terrain', 'coastal_water', 'agriculture', 'water']]
plt.figure(figsize=(18,18))
seaborn.heatmap(df_seaborn_heatmap.corr(), annot=True, cmap="coolwarm").set_title('Seaborn')
```

```python
data_norm.columns

data_norm_new=data_norm[['observer_intensity','block', 'woy','precipitation', 'temperature',
       'pop001', 'pop009', 'pop025', 'pop049', 'pop081', 'pop121', 'pop169',
       'pop225', 'municipality', 'province', 'water_authority', 'rijk',
       'other_maintainers', 'roadlength', 'built_up', 'dry_natural_terrain',
       'greenhouse_horticulture', 'railway', 'nature_reserves', 'company_site',
       'wet_natural_terrain', 'airport', 'forest', 'coastal_water',
       'semi_built', 'agriculture', 'recreation', 'water', 'highway']]
```

```python
df=data_norm_new.dropna()
pnr_value = pd.DataFrame (columns=['r','p'])

for col in data_norm_new:
    if pd.api.types.is_numeric_dtype(df[col]):
        r , p =stats.pearsonr (df.observer_intensity, df[col])
        pnr_value.loc[col]=[round(r,3), round(p,9)]
pnr_value
```

```python
df_seaborn_heatmap=pnr_value
plt.figure(figsize=(18,18))
seaborn.heatmap(df_seaborn_heatmap, annot=True, cmap="coolwarm")
```
